```python
import dotenv
_ = dotenv.load_dotenv()
```

```python
SYSTEM_PROMPT = """
Du bist ein KI-Assistent, der Rollenspielern bei der Erstellung ihrer Character Sheets hilft.

Der Nutzer wird dir eine kurze Beschreibung des Charakters geben, den du erstellen sollst.
Gib ein vollständiges und sinnvoll unterteiltes Character Sheet zurück.

Du spezialisierst dich auf das klassische Rollenspiel "Dungeons and Dragons" (DND) und nimmst
standardmäßig an, dass der Nutzer ein DND Spieler ist. Du kennst allerdings auch sämtliche
weiteren Rollenspiele und kannst flexibel Character Sheets für jedes Rollenspiel erstellen,
das der Nutzer spielt.

Wähle basierend auf dem Spiel, welche Punkte in dem Character Sheet enthalten sind.
Überlege dir zuerst die groben Rahmendaten wie z.B. Name, Rasse und Klasse des Characters
und gehe daraufhin in die Einzelheiten ein.

Gib NUR das Charactersheet zurück. Gib keine einleitenden oder ableitenden Kommentare oder
Informationen zurück, die nicht in einem Charactersheet enthalten wären.
""".strip()
```

```python
import openai

def generate_character(description: str) -> str:
    response = openai.chat.completions.create(
        model="gpt-4-0125-preview",
        messages=[
            {
                "role": "system",
                "content": SYSTEM_PROMPT,
            },
            {
                "role": "user",
                "content": description,
            },
        ],
    )

    return response.choices[0].message.content
```

```python
from rich.console import Console
from rich.markdown import Markdown

def print_markdown(text: str) -> None:
    markdown = Markdown(text)
    console = Console()
    console.print(markdown)
```

```python
character = generate_character("Ein Zwerg, der 3 Meter groß ist und fliegen kann")
print_markdown(character)
```

```python
character = generate_character("Ein kurzbeiniger Kämpfer, der Feinde mit Pfannen bewirft.")
print_markdown(character)
```
